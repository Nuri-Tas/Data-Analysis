# Basics 

```python
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.sinc(x)

plt.plot(x, y);
```

### Scatter Plot 

```python
x = np.random.randn(30) + 3
y = np.random.rand(len(x)) + 3

sizes = np.random.uniform(15, 90, len(x))
colors = np.random.uniform(15, 90, len(x))

fig, ax = plt.subplots()

ax.scatter(x, y, sizes=sizes, c=colors, vmin=10, vmax=100);
```

### Bar

```python
x = np.arange(0, 11)
y = np.random.uniform(3, 15, len(x))

fig, ax = plt.subplots(figsize=(12,5))
ax.bar(x, y)

ax.set(xticks=np.arange(0, 11), yticks=np.arange(0,15));
```

### Step Plot

```python
x = np.linspace(0,8,12)
y = np.random.uniform(2,6, len(x))

fig, ax = plt.subplots(figsize=(12, 5))

ax.step(x, y, lw=2.5, label='performance')
ax.step(x, y**1/2 - y, lw=2.5, label='loss')

ax.legend(loc=3);
```
